## 1 Filters methods

### 1.1 Filter methods - constant, quasi constant and duplicated features

##### 1.1.1 Constant features

```python
# To remove constant features we will use Feature-engine
sel = DropConstantFeatures(tol=1, variables=None, missing_values='raise')

sel.fit(X_train_c)
```

```python
# number of constant features

len(sel.features_to_drop_)
```

##### 1.1.2 Quasi constant features

```python
# To remove quai constant features we will use Feature-engine
sel = DropConstantFeatures(tol=0.998, variables=None, missing_values='raise')
# we set the tolerance in 0.998

sel.fit(X_train_c)
```

```python
# number of quasi constant features

len(sel.features_to_drop_)
```

##### 1.1.2 Duplicate features

```python
# To remove duplicate features we will use Feature-engine
# set up the selector
sel = DropDuplicateFeatures(variables=None, missing_values='raise')

# find the duplicate features, this might take a while
sel.fit(X_train_c)
```

```python
len(sel.features_to_drop_)
```

##### 1.1.3 Correlation SmartCorrelationSelection - Model perfomance

We will keep a feature from each correlation group based on the performance of a random forest.

We considered to Spearman's Rank. We think that Spearman will be more efficient because can handle both continuous and ordinal varibles. Both types of variables are in our dataset.

```python
# random forest
rf = RandomForestClassifier(
    n_estimators=10,
    random_state=0,
    n_jobs=4,
)

# correlation selector
sel = SmartCorrelatedSelection(
    variables=None, # if none, selector examines all numerical variables
    method="spearman",
    threshold=0.8,
    missing_values="raise",
    selection_method="model_performance",
    estimator=rf,
    scoring="f1_macro",
    cv=3,
)

# this may take a while, because we are training
# a random forest per correlation group

sel.fit(X_train_c, y_train_c)
```

```python
sel.correlated_feature_sets_
```

#### 1.1.4 Correlation SmartCorrelationSelection - Variance

```python
sel = SmartCorrelatedSelection(
    variables=None,
    method="spearman",
    threshold=0.8,
    missing_values="raise",
    selection_method="variance",
    estimator=None,
    scoring="f1_macro",
    cv=3,
)

sel.fit(X_train_c, y_train_c)
```

```python
sel.correlated_feature_sets_
```

### 1.2 Pipeline filter methods

#### 1.2.1 Pipeline feature selection

```python
# Define your random forest estimator
rf = RandomForestClassifier(
    n_estimators=10,
    random_state=0,
    n_jobs=4,
)
# we stack all the selection methods inside a pipeline
pipe_model_performance = Pipeline([
    ('constant', DropConstantFeatures(tol=0.998)),
    ('duplicated', DropDuplicateFeatures()),
    ('correlation_model_performance', SmartCorrelatedSelection(
        method='spearman',
        missing_values='raise',
        scoring='f1_macro',
        selection_method='model_performance',
        estimator=rf,
        cv=3,
    )),
])
pipe_model_performance.fit(X_train_c, y_train_c)
```

#### 1.2.2 Pipeline variance

```python
# we stack all the selection methods inside a pipeline

pipe_variance = Pipeline([
    ('constant', DropConstantFeatures(tol=0.998)),
    ('duplicated', DropDuplicateFeatures()),
    ('correlation_variance', SmartCorrelatedSelection(
        method='spearman',
        missing_values='raise',
        scoring='f1_macro',
        selection_method='variance'
    )),
])
pipe_variance.fit(X_train_c)
```

```python
# X_train
X_train_performance = pipe_model_performance.transform(X_train_c)
X_train_variance = pipe_variance.transform(X_train_c)

#X_test
X_test_performance = pipe_model_performance.transform(X_test_c)
X_test_variance = pipe_variance.transform(X_test_c)

print('X_train_c', X_train_performance.shape, X_train_variance.shape, '\n',
      'X_test_c', X_test_performance.shape, X_test_variance.shape )

```

## 2 Models peformance

```python
#Function to evaluate the perfomance

classifiers = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Extra Trees': ExtraTreesClassifier(n_estimators=100, random_state=42),
    'XGBoost': XGBClassifier()
}
def evaluate_performance(X_train_dataset, X_test_dataset, y_train_dataset, y_test_dataset, classifiers):
    results = {}
    for name, clf in classifiers.items():
        clf.fit(X_train_dataset, y_train_dataset)

        # Train set predictions
        train_preds = clf.predict(X_train_dataset)
        train_f1 = f1_score(y_train_dataset, train_preds, average='macro')

        # Test set predictions
        test_preds = clf.predict(X_test_dataset)
        test_f1 = f1_score(y_test_dataset, test_preds, average='macro')

        results[name] = {'Train F1 Macro': train_f1, 'Test F1 Macro': test_f1}
    return results

```

```python
#Iterate over each dataset and evaluate

datasets = {
    'Performance 2': (X_train_pefomance_2_pl, X_test_pefomance_2_pl, y_train_2c, y_test_2),
    'Variance 2': (X_train_variance2_pl, X_test_variance2_pl,  y_train_2c, y_test_2),
    'Tree 2': (X_train_tree2_pl, X_test_tree2_pl,  y_train_2c, y_test_2),
    'Tree_perfomance 2': (X_train_performance_tree2_pl, X_test_performance_tree2_pl,  y_train_2c, y_test_2),
    'Tree_variance 2': (X_train_variance_tree2_pl, X_test_variance_tree2_pl,  y_train_2c, y_test_2),

    'Performance 2a': (X_train_pefomance_2a_pl, X_test_pefomance_2a_pl, y_train_2ac, y_test_2a),
    'Variance 2a': (X_train_variance2a_pl, X_test_variance2a_pl,y_train_2ac, y_test_2a),
    'Tree 2a': (X_train_tree2a_pl, X_test_tree2a_pl,y_train_2ac, y_test_2a),
    'Tree_perfomance 2a': (X_train_performance_tree2a_pl, X_test_performance_tree2a_pl,y_train_2ac, y_test_2a),
    'Tree_variance 2a': (X_train_variance_tree2a_pl, X_test_variance_tree2a_pl,y_train_2ac, y_test_2a),

    'Performance 3': (X_train_pefomance_3_pl, X_test_pefomance_3_pl,y_train_3c, y_test_3),
    'Variance 3': (X_train_variance3_pl, X_test_variance3_pl,y_train_3c, y_test_3),
    'Tree 3': (X_train_tree3_pl, X_test_tree3_pl,y_train_3c, y_test_3),
    'Tree_perfomance 3': (X_train_performance_tree3_pl, X_test_performance_tree3_pl,y_train_3c, y_test_3),
    'Tree_variance 3': (X_train_variance_tree3_pl, X_test_variance_tree3_pl,y_train_3c, y_test_3)
    }



final_results = {}
for dataset_name, (X_train_dataset, X_test_dataset, y_train_dataset, y_test_dataset) in datasets.items():
    print(f"Evaluating {dataset_name}...")
    results = evaluate_performance(X_train_dataset, X_test_dataset, y_train_dataset, y_test_dataset, classifiers)
    final_results[dataset_name] = results
```

```python
# Print the results
for dataset_name, classifiers_results in final_results.items():
    print(f"\nResults for {dataset_name}:")
    for clf_name, scores in classifiers_results.items():
        print(f"{clf_name} - Train F1 Macro: {scores['Train F1 Macro']:.4f}, Test F1 Macro: {scores['Test F1 Macro']:.4f}")
```

Observations based on the results:

Overfitting: Both Random Forest and Extra Trees classifiers show signs of overfitting, as indicated by the near-perfect F1 scores on the training data and significantly lower scores on the test data. XGBoost, on the other hand, seems to generalize better to the test data, as the gap between its training and test scores is smaller.


Best Model: While XGBoost has a lower tendency to overfit, its test performance is not always the highest among the three classifiers. However, given its better generalization, it might be a safer choice in a real-world scenario where the data distribution can change.


Feature Selection: The results vary based on the feature selection method used (Performance, Variance, Tree). It's essential to understand the nature of these feature selection methods and choose the one that makes the most sense for your specific problem.

### 4.1 K Neighbours Classifier and Decision Tree Classifier

```python
#Function to evaluate the perfomance

classifiers = {
     'K Neighbors': KNeighborsClassifier(n_neighbors=5),
    'Decision Tree': DecisionTreeClassifier(random_state=0)
}
def evaluate_performance(X_train_dataset, X_test_dataset, y_train_dataset, y_test_dataset, classifiers):
    results = {}
    for name, clf in classifiers.items():
        clf.fit(X_train_dataset, y_train_dataset)

        # Train set predictions
        train_preds = clf.predict(X_train_dataset)
        train_f1 = f1_score(y_train_dataset, train_preds, average='macro')

        # Test set predictions
        test_preds = clf.predict(X_test_dataset)
        test_f1 = f1_score(y_test_dataset, test_preds, average='macro')

        results[name] = {'Train F1 Macro': train_f1, 'Test F1 Macro': test_f1}
    return results
```

```python

datasets = {
    'Performance 2': (X_train_pefomance_2_pl, X_test_pefomance_2_pl, y_train_2c, y_test_2),
    'Performance 2a': (X_train_pefomance_2a_pl, X_test_pefomance_2a_pl, y_train_2ac, y_test_2a),
    'Variance 2a': (X_train_variance2a_pl, X_test_variance2a_pl,y_train_2ac, y_test_2a),

    }

final_results = {}
for dataset_name, (X_train_dataset, X_test_dataset, y_train_dataset, y_test_dataset) in datasets.items():
    print(f"Evaluating {dataset_name}...")
    results = evaluate_performance(X_train_dataset, X_test_dataset, y_train_dataset, y_test_dataset, classifiers)
    final_results[dataset_name] = results
```

```python
#Function to evaluate the perfomance
classifiers = {
    'K Neighbors': KNeighborsClassifier(n_neighbors=5),
    'Decision Tree': DecisionTreeClassifier(random_state=0)
}

def evaluate_performance(X_train_dataset, X_test_dataset, y_train_dataset, y_test_dataset, classifiers):
    results = {}
    for name, clf in classifiers.items():

        # Apply Min-Max scaling if the classifier is K Neighbors
        if isinstance(clf, KNeighborsClassifier):
            scaler = MinMaxScaler()
            X_train_dataset = scaler.fit_transform(X_train_dataset)
            X_test_dataset = scaler.transform(X_test_dataset)

        clf.fit(X_train_dataset, y_train_dataset)

        # Train set predictions
        train_preds = clf.predict(X_train_dataset)
        train_f1 = f1_score(y_train_dataset, train_preds, average='macro')

        # Test set predictions
        test_preds = clf.predict(X_test_dataset)
        test_f1 = f1_score(y_test_dataset, test_preds, average='macro')

        results[name] = {'Train F1 Macro': train_f1, 'Test F1 Macro': test_f1}
    return results

datasets = {
    'Performance 2': (X_train_pefomance_2_pl, X_test_pefomance_2_pl, y_train_2c, y_test_2),
    'Performance 2a': (X_train_pefomance_2a_pl, X_test_pefomance_2a_pl, y_train_2ac, y_test_2a),
    'Variance 2a': (X_train_variance2a_pl, X_test_variance2a_pl,y_train_2ac, y_test_2a),
}

final_results = {}
for dataset_name, (X_train_dataset, X_test_dataset, y_train_dataset, y_test_dataset) in datasets.items():
    print(f"Evaluating {dataset_name}...")
    results = evaluate_performance(X_train_dataset, X_test_dataset, y_train_dataset, y_test_dataset, classifiers)
    final_results[dataset_name] = results
```

```python
# Print the results
for dataset_name, classifiers_results in final_results.items():
    print(f"\nResults for {dataset_name}:")
    for clf_name, scores in classifiers_results.items():
        print(f"{clf_name} - Train F1 Macro: {scores['Train F1 Macro']:.4f}, Test F1 Macro: {scores['Test F1 Macro']:.4f}")
```
